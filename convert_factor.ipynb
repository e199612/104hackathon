```python
# -*- coding: utf-8 -*-
%matplotlib inline
#import datetime
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import re,math
import pandas as pd
from math import isnan
import gc
```

```python
# load testdata
pop=pd.read_csv('../data/user_log_testset/user_log_testset.csv',sep='|')
```

```python
pop_filter=pop[pop['action']!='viewCust']
pop_filter.head()
```

```python
count_apply=pop_filter[['uid','jobNo','action','invoice']].groupby(['uid','jobNo','action']).count().unstack()
```

```python
count_apply.head()
```

```python
turn_uid=count_apply.reset_index('uid')
turn_jobNo=turn_uid.reset_index('jobNo')
```

```python
turn_jobNo
```

```python
final=pd.DataFrame({'uid':turn_jobNo.uid,'jobNo':turn_jobNo.jobNo, 'saveJob':turn_jobNo.invoice.saveJob, 'viewJob':turn_jobNo.invoice.viewJob})
```

```python
final_pivot=final[['uid','jobNo','saveJob','viewJob']]
```

```python
final_pivot['jobNo']=final_pivot['jobNo'].apply(np.int64)
```

```python
final_pivot.to_csv('../data/user_log_testset/test_pivot_user_log.csv',sep='|')
```
